```python
import os
import numpy as np
import cv2
def calc_light(image):
    b,g,r,t = cv2.mean(cv2.imread(image))
    return (b+g+r)/3
def fusion(image,event):
    light = calc_light(image)
    print(light)
    if light<110 and light >60:
        os.system("python mmyolo/projects/easydeploy/tools/image-demo.py \
        {} \
        mmyolo/configs/yolov5/yolov5_s-v61_fast_1xb12-40e_image.py \
        image.onnx \
        --device cpu".format(image))
    else:
        os.system("python mmyolo/projects/easydeploy/tools/image-demo.py \
        {} \
        mmyolo/configs/yolov5/yolov5_s-v61_fast_1xb12-40e_cat.py \
        event.onnx \
        --device cpu".format(event))

fusion("testimage/image1.png","testimage/event1.png")
fusion("testimage/image2.png","testimage/event2.png")
fusion("testimage/image3.png","testimage/event3.png")
```
